```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib notebook

plt.style.use('seaborn')
```

```python
mean = pd.read_csv('Mean.csv')
mean.head()
```

```python
minimum = pd.read_csv('Minimum.csv')
minimum.head()
```

```python
rainfall = pd.read_csv('Rainfall.csv')
rainfall.head()
```

```python
weather = pd.DataFrame(mean['YEAR'])
weather['Mean'] = mean['ANNUAL']
weather.head()
```

```python
weather['Minimum'] = minimum['ANNUAL']
weather.head()
```

```python
filt = rainfall['YEAR'] <= 2012
weather['Rainfall'] = rainfall.loc[filt, 'ANN'] / 100
weather = weather[weather['YEAR'] > 1980]
weather = weather.set_index('YEAR')
weather.head()
```

```python
x = weather.index.tolist()
y1 = weather['Mean'].tolist()
y2 = weather['Minimum'].tolist()
y3 = weather['Rainfall'].tolist()

plt.xticks(x, rotation=40)
plt.tick_params(axis="x", labelsize=8)

plt.plot(x, y1, '-s', label = "Mean Temperature ($\,^{\circ}\mathrm{C}$)")
plt.plot(x, y2, '-o', label = "Minimum Temperature ($\,^{\circ}\mathrm{C}$)")
plt.plot(x, y3, '-*', label = "Rainfall (0.1cm)")

plt.xlabel("Years")
plt.ylabel("")
plt.title('Average Mean and Minimum Temperatures in India(1981-2012) (in $\,^{\circ}\mathrm{C}$) and Rainfall')
 
plt.legend(loc="best")
```
